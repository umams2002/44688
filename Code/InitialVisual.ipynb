### check heat map 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("eda_processed_adult.csv")
df.head()
```

### Income Distribution by Age

```python
### Income Distribution by Age
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='income', y='age')
plt.title('Boxplot of Age by Income')
plt.show()
```

### Hours-per-Week Distribution by Income

```python
### Hours-per-Week Distribution by Income
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='income', y='hours-per-week')
plt.title('Boxplot of Hours-per-Week by Income')
plt.show()
```

### Pair Plot

```python
### Pair Plot
sns.pairplot(df, hue='income')
plt.suptitle('Pair Plot of Features Colored by Income', y=1.02)
plt.show()
```

###  Bar Plot of Education Level by Income

```python
###  Bar Plot of Education Level by Income
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='education', hue='income')
plt.title('Education Level by Income')
plt.xticks(rotation=90)
plt.show()
```

### Correlation Heatmap

```python
### Correlation Heatmap
numerical_df = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = numerical_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
