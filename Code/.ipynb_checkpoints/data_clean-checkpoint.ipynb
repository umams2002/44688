### Import pandas and read CSV file

```python
import pandas as pd
import numpy as np

data = pd.read_csv("adult.csv")
data.columns
data.describe()
```

### clean any duplicates value

```python
data.drop_duplicates(inplace =True)
data.describe()
```

### clean any missing values

```python
data.replace("?", np.nan, inplace=True)
data.describe()
```

### Include new column to define row count

```python
data['RowCount'] = data.index+1
```

### cleaned data into extracted file in csv

```python
data.to_csv("processed_adult.csv", index=False)
```
